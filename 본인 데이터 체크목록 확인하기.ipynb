```python
tmp = pd.Series(dirlst).apply(lambda x : x.split()[0] if len(x.split())>1 else np.nan).dropna().values
```

```python
#1 첫번째 방법
# 맨앞에 있는 키워드를 공통으로 가지고 있는 녀석들끼리 만 묶어서 그룹으로 만든경우
# 중복없음
# 단점은 앞에서 다 가져가버리면 없는놈이 있음 
# (베베마루 라는 키워드를 가지고 있는 술들이 있는데 로제 라는 키워드가 먼저 나오면서 다 가져가버려서 구분이 안감)

groups = []
dirlst = os.listdir(origin_path)
for ttmp in tmp:
    if len(ttmp) == 1:
        continue
    group = []
    ind = 0
    length = len(dirlst)
    while length>ind:
        if ttmp in dirlst[ind]:
            group.append(dirlst.pop(ind))
            length-=1
        else:
            ind+=1
    groups.append((ttmp,group))
groups
```

```python
# 두번째 방법
# 앞에거처럼 맨 앞 키워드를 가지고 있으면 죄다 그룹화함
# 중복까지 다나옴
# 단점 중복이 너무 많음, 순서까지 중복이라 참고하기 좀 빡셀듯

groups = dict()
dirlst = os.listdir(origin_path)
for ttmp in tmp:
    if len(ttmp) == 1:
        continue
    group = set()
    for dr in dirlst:
        if ttmp in dr:
            group.add(dr)
    groups[ttmp]=list(group)
groups
```

```python
# 3번째 방법
# 이게 그나마 제일 보기 좋지 않을까?
# 특정 주류 이름에 있는 모든 split들 다 한번씩 확인하는 코드
# 연관성 있는것으로 추정되는애들 다 체크 가능
# 순서간의 상관관계가 없어 좀 보기 빡셀수 있음
checklst = []
for keyword in dirlst:
    keys = keyword.split()
    groups = set()
    for key in keys:
        if len(key)<2:
            continue
        group = []
        for dr in dirlst:
            if key in dr:
                group.append(dr)
            if len(group)>10:
                group=[]
                break
        for elem in group:
            groups.add(elem)
    checklst.append([keyword,groups])
checklst
```
