```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
import tensorflow as tf 

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('../kdt4/workplace2/data/deep_data_files-20220321T014240Z-001/deep_data_files/fish.csv')
df.head()
```

```python
sns.pairplot(df,hue='Species')
```

```python
y = df.Species # y 
x = df.drop('Species',axis=1)
```

```python
df.Species.value_counts() # 
```

```python
y
```

```python
x
```

```python
e = LabelEncoder() # 
y_en = e.fit_transform(y)
y_en.shape
```

```python
y_oh = tf.keras.utils.to_categorical(y_en) #      sparse 
y_oh
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
x_sc = scaler.fit_transform(x)
x_sc
```

```python
x_sc.shape #5
```

```python
model = Sequential()
model.add(Dense(128, input_dim=5,activation='relu'))
model.add(Dense(64,activation='relu')) #
model.add(Dense(32,activation='relu')) # 
model.add(Dense(7,activation='softmax')) # 7   7  
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer='adam') # 
```

```python
hist = model.fit(x_sc,y_oh,epochs=100,batch_size=1) #     () batchsize
```

```python
model.evaluate(x_sc,y_oh)
```

```python
model.predict(x_sc).argmax(axis=1)
```
