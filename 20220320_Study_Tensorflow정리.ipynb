```python
# # 텐서플로우에선 딥러닝이 어떻게 이루어지나?
# # 딥러닝으로 간단한 수학문제 풀어보기
# # 문제 키와 신발사이즈는 어떤 관련이 있을까요 ?
# 키 = [170,180,175,160]
# 신발 = [260,270,265,255]
# # ??? = a 170 +b  직선의 방정식.. 키를알면 신발사이즈를 알수있음

키 = 170
신발 = 260 

# 신발 = 키 * a + b  키랑 신발이 뭔가연관이 되어있을거같은데 미지수를 곱해서 예측함

a = tf.Variable(0.1)
b = tf.Variable(0.2) 

# def 손실함수():
#     return tf.square손실값(실제값 - 예측값)  :^2이렇게 하진않고 tf.square하면댐

def 손실함수():
    예측값 = 키 *a+b
    return tf.square(260 - 예측값)

opt = tf.keras.optimizers.Adam(learning_rate=0.1) # 경사하강법을 이용해서 w변수들을 업데이트해주는 고마운 친구 
# gradient를 알아서 스마트하게 바꿔줌 Adam..  learning_rate지정도 가능
# opt.minimize(손실함수,var_list=[a,b]) #경사하강법으로 업데이트할 weight Variable목록
# ↑ 경사하강 1번해줌 == a,b를 1번 수정해줌 a와b의값이 0.1, 0.2였었는데 실제답과 유사한 a,b값으로 변한다는뜻.
# 한번만 하면 제대로 예측하는값이 나오지않음.. 그래서 여러번 반복하면됨. 최적값이 나올때까지 

for i in range(300):
    opt.minimize(손실함수,var_list=[a,b])
    print(a.numpy(),b.numpy()) # 보기쉽게 넘파이 
    
# 예측한 신발사이즈 = 키*1.52+1.62 
print(키*1.52+1.62) 

# 우리가만든 linear regression 모델임. 이제 키만 알면 신발사이즈를 추정할 수 있는것.
```

```python
# 대학 합격 예측 AI 만들기 
# 붙을까요 ?X 붙을확률구하기 O
import pandas as pd
import numpy as np
data = pd.read_csv('./gpascore.csv')
# 데이터 전처리 해야함. nan값에 평균값을 넣거나 행을 삭제하거나 
data.dropna()

y데이터 = data['admit'].values # y입력값 만들기 ..걍 리스트로 다 출력해줌.
x데이터 = []

for i, rows in data.iterrows(): #data라는 데이터프레임을 가로한줄씩 출력해라.. 
    x데이터.append([rows['gre'],rows['gpa'],rows['rank']])
# 근데 .. 파이썬 리스트를 그대로 집어 넣을 순 없음.. 그래서 numpy array 나 tensor를 넣어야함.



model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64,activation='relu'), # 노드의 갯수
    tf.keras.layers.Dense(128,activation='relu'), # 히든레이어의 갯수 결과가 잘 나올때까지 실험으로 파악해야함. 근데 2의제곱으로 많이 표현함 
    tf.keras.layers.Dense(1, activation='sigmoid'), # 마지막출력레이어는 1개만 와야하지 않겠뉘.. 0과1로 나타나니까 .. 근데 여러가지로 예측하려면 마지막 레이어를 신경써서 만들면됨.(확률한개를 예측할때도 1개)
]) #<= 딥러닝 모델 만드는법

# 레이어를 디자인할때 activation function 이 꼭 필요함.
# 마지막 레이어는 예측결과를 뱉어야함 . 0~1사이의 확률로 (0~1사이의 확률을 뱉고싶으면 sigmoid)
# 마지막 출력결과가 0.9 =입학확률 90%

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
# optimizer 경사하강법으로 수정함.. 기울기를 뺄때 lerning rate 곱해서 빼준다했는데 optimizer는 그 값을 알맞게 조정해줌. 
# loss function은 확률을 구하고싶을때 binary_crossentropy <- 결과가 0과1사이의 분류/확률문제에서 씀.. 
# metrics는 모델평가할때 모쓸지 

model.fit(np.array(x데이터),np.array(y데이터),epochs=10) #학습시키기 , epochs 몇번 학습을 시킬까

# [[380,3.21,3],[660,3.67,3]] x데이터 이런식으로 만들어야함
# [0,1,1,0...] y라벨 위로가서 pandas부터 

# 예측 
model.predict([[750,3.70,3],[400,2.2,1]]) # 예측은 model.predict(테스트해볼x데이터)

# 1.모델만들기
# 2.데이터집어넣고 학습
# 3.새로운 데이터 예측
# +데이터 전처리
# +모델 튜닝 
# 딥러닝 학습은 연구과정임.. => 실험이 매우 중요 
```
