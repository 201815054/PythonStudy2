```python
mobile = tf.keras.applications.mobilenet.MobileNet()
```

```python
def prepare_image(file):
    img_path = '../data/cityscapes_data/train/'
    img = image.load_img(img_path+file,target_size=(256,512))
    img_array = image.img_to_array(img)
    img_array_expanded_dims =np.expend_dims(img_array,axis=0)
    return tf.keras.applications.mobilenet.MobileNet.preprocess_input(img_array_expanded_dims)
```
