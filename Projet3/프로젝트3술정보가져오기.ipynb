```python
from selenium import webdriver
from bs4 import BeautifulSoup 
import time 
import re
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./더술_전통주리스트.csv',index_col=0)
df.head()
```

```python
driver = webdriver.Chrome()
```

```python
# 1부터 1000까지 반복해야하는데 PR000000{}이 뒤에 100이하면 앞에 0하나 붙여야됨.
# i가 100 이하면 url에 정규표현식으로 0을 하나 붙임.. 
name_lst=[]
kinds_lst = []
material_lst = []
alchol_level_lst = []
ml_lst = []
intro_lst = []
food_lst =[]
home_lst =[]
addr_lst=[]
homep_lst=[]
ask_lst=[]


for i in range(1,1000):
    
    
    url = 'https://thesool.com/front/find/M000000082/view.do?searchType=2&searchKey=&searchKind=&levelType=&searchString=&productId=PR00000{:03}&pageIndex=2&categoryNm=&kind='.format(i)
    print(url)
    driver.get(url)
    time.sleep(1)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    name = soup.select('#content > div > div > div.product-view > dl > dt')
    if name:
        name_lst.append(name[0].text.strip('\n\t'))        
    else:
        continue
    
    kinds = soup.select('#content > div > div > div.product-view > dl > dd.detail > div.info > ul > li:nth-child(1)>span')
    if kinds and kinds[0].text.strip('\n\t'):
        kinds_lst.append(kinds[0].text.strip('\n\t')) # 311번부터 패스값을 text형식으로 뽑을 수 없음.. 패스는 맞으니 if문이 구동하는데 
    else:
        kinds_lst.append(np.nan)
        
    material = soup.select('#content > div > div > div.product-view > dl > dd.detail > div.info > ul > li:nth-child(2)>span')
    if material and material[0].text.strip('\n\t'):
        material_lst.append(material[0].text.strip('\n\t'))
    else:
        material_lst.append(np.nan)
        
    alchol_level = soup.select('#content > div > div > div.product-view > dl > dd.detail > div.info > ul > li:nth-child(3)>span')
    if alchol_level and re.search('(\d)*%',alchol_level[0].text):
        alchol_level_lst.append(re.search('(\d)*%',alchol_level[0].text)[0])
    else:
        alchol_level_lst.append(np.nan)
        
    ml = soup.select('#content > div > div > div.product-view > dl > dd.detail > div.info > ul > li:nth-child(4)>span')
    if ml and ml[0].text.strip('\n\t'):
        ml_lst.append(ml[0].text.strip('\n\t'))
    else:
        ml_lst.append(np.nan)
        
    intro = soup.select('#content > div > div > div.product-view > dl > dd.intro > div.text')
    if intro and intro[0].text.strip('\n\t'):
        intro_lst.append(intro[0].text.strip('\n\t'))
    else:
        intro_lst.append(np.nan)
        
    food = soup.select('#content > div > div > div.product-view > dl > dd.food > div.text')
    if food and food[0].text.strip('\n\t'):
        food_lst.append(food[0].text.strip('\n\t'))
    else:
        food_lst.append(np.nan)
        
    home = soup.select('#content > div > div > div.product-view > dl > dd.place > div.text > ul > li:nth-child(1)>span')
    if home and home[0].text.strip('\n\t'):
        home_lst.append(home[0].text.strip('\n\t'))
    else:
        home_lst.append(np.nan)
        
    addr = soup.select('#content > div > div > div.product-view > dl > dd.place > div.text > ul > li:nth-child(2)>span')
    if addr and addr[0].text.strip('\n\t'):
        addr_lst.append(addr[0].text.strip('\n\t'))
    else:
        addr_lst.append(np.nan)
        
    homep = soup.select('#content > div > div > div.product-view > dl > dd.place > div.text > ul > li:nth-child(3)>span')
    if homep and homep[0].text.strip('\n\t'):
        homep_lst.append(homep[0].text.strip('\n\t'))
    else:
        homep_lst.append(np.nan)
        
    ask = soup.select('#content > div > div > div.product-view > dl > dd.place > div.text > ul > li:nth-child(4)')
    if ask and ask[0].text.strip('\n\t'):
        ask_lst.append(ask[0].text.strip('\n\t'))
    else:
        ask_lst.append(np.nan)
        
liqinfo = pd.DataFrame({'name':name_lst,
                   'kinds':kinds_lst,
                   'material':material_lst,
                   'alchol':alchol_level_lst,
                   'ml':ml_lst,
                   'intro':intro_lst,
                   'food':food_lst,
                   'home':home_lst,
                   'addr':addr_lst,
                   'homep':homep_lst,
                   'ask':ask_lst})
liqinfo

# 술정보csv파일 = 술정보.to_csv('LiquorInfo.csv',encoding='UTF-8')

# product number가 1000이상이 넘어갔을때 제품이 없는것 확인 완료 
# 가장먼저 나오는 이름이 안나오면 바로 다음페이지로 넘어감 (페이지를찾을수없습니다 페이지 거르기)
# 컬럼추출이 불가능할때 ( nontype으로 들어가있을때) nan값 채워주기
```
